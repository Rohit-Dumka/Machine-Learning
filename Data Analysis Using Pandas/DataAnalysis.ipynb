sometimes missing values  are encoded in different ways. They can appear as NaN, NA, ?, zeros, xx, -1 or a blank space ''.

So it is essential that we should first convert all the other values(?,zeroes,NA etc ) to NaN.

If the missing vlaues isn't identified as NaN, then we have to first convert or replace such non NaN entry with a NaN.

Dataframe also has an **isin()** method.
When calling isin,we pass a set of values as either an array or dict. if values is in array,isin returns a DataFrame of booleans that is the same shape as the original DataFrame, with True wherever the element is in the sequence of vlaues.

```python
#Let's use isin function for searching row with given values

values=[1006038,'P00375436','F','55+',1,'C','2',0,20,2.0,11.0,365]

df1_indexed=df1.isin(values)
df1_indexed
```

**Find the total amoutn generate via website by selling product**

We can apply aggregation on a single column of a dataframe

```python
#we can use np.sum aggregation to get total purchased amount

df1['Purchase'].aggregate(np.sum)
#gives sum of purchase over whole dataset
```

we can also apply multiple functions on a single column of a dataframe

```python
#find sum and mean value after doing aggregation over purchase column

df1['Purchase'].aggregate([np.sum,np.mean])
```

```python
# find mean for 'Product_Category_1' , 'Product_Category_2', 'Product_Category_3'

df1[['Product_Category_1','Product_Category_2','Product_Category_3']].aggregate(np.mean)
```

we can also apply multiple functions on multiple columns of a dataframe

```python
#find mean and sum for 'Product_Category_1' , 'Product_Category_2', 'Product_Category_3'

df1[['Product_Category_1','Product_Category_2','Product_Category_3']].aggregate([np.sum,np.mean])
```
